```python
import py3Dmol #почитать про эту библиотеку.. вероятно, в ней есть смысл, ибо в одном месте я её нашёл.
```

```python
with open("/content/6hh0.pdb1") as ifile:
    system = "".join([x for x in ifile])
```

```python
view = py3Dmol.view(width=400, height=300)
view.addModelsAsFrames(system)
view.setStyle({'model': -1}, {"cartoon": {'color': 'spectrum'}})
view.zoomTo()
view.show()
```
